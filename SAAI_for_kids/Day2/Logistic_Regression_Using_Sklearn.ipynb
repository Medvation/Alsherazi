Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
```

Create Data

```python
# Data: Hours studied and corresponding pass/fail (1/0)
hours_studied = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
passed = np.array([0, 0, 1, 1, 1])  # 0 = Fail, 1 = Pass
```

Import Logistic Regression model from Sklearn

```python
# Initialize and train the Logistic Regression model
log_reg = LogisticRegression()
```

Train the nodel

```python
log_reg.fit(hours_studied, passed)
```

Make Predictions

```python
# Test the model prediction for 3 hours of study
predicted_probability = log_reg.predict_proba([[3]])[0][1]  # Probability of passing
predicted_class = log_reg.predict([[3]])[0]  # Class prediction (0 or 1)

print(f"Predicted probability of passing for 3 hours of study: {predicted_probability}")
print(f"Predicted class for 3 hours of study: {predicted_class}")
```

Visualization

```python
# Visualization
# Generate a range of hours studied for the curve
hours_range = np.linspace(0, 6, 100).reshape(-1, 1)
predicted_probabilities = log_reg.predict_proba(hours_range)[:, 1]  # Probability of passing for each hour value

# Plot the logistic regression curve
plt.figure(figsize=(10, 6))
plt.plot(hours_range, predicted_probabilities, label="Logistic Regression Curve", color="orange")

# Plot the original data points (pass/fail)
plt.scatter(hours_studied, passed, color='blue', label="Data (0 = Fail, 1 = Pass)", s=100)

# Labels and title
plt.axhline(0.5, color='red', linestyle='--', label="Decision Boundary (0.5)")
plt.xlabel("Hours Studied")
plt.ylabel("Probability of Passing")
plt.title("Logistic Regression: Probability of Passing Based on Hours Studied")
plt.legend()

# Show the plot
plt.show()
```
