Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

Create Data

```python
# Data: Hours studied, Assignments completed, and Test scores
X = np.array([[1, 1],  # 1 hour, 1 assignment
              [2, 2],  # 2 hours, 2 assignments
              [3, 1],  # 3 hours, 1 assignment
              [4, 3],  # 4 hours, 3 assignments
              [5, 4]]) # 5 hours, 4 assignments

y = np.array([50, 60, 70, 80, 90])  # Test scores
```

Import Linear Regression model from Sklearn

```python
# Initialize and train the Linear Regression model
lin_reg = LinearRegression()
```

Train the nodel

```python
lin_reg.fit(X, y)
```

Make Predictions and classification

```python
# Test the model prediction for 3 hours studied and 2 assignments completed
predicted_score = lin_reg.predict([[3, 2]])[0]
print(f"Predicted score for 3 hours of study and 2 assignments completed: {predicted_score}")
```

Visualization

```python
# Create a grid for hours studied and assignments completed
hours_range = np.linspace(1, 5, 5)
assignments_range = np.linspace(1, 4, 5)
hours_grid, assignments_grid = np.meshgrid(hours_range, assignments_range)
X_grid = np.c_[hours_grid.ravel(), assignments_grid.ravel()]

# Predict test scores for all combinations of hours and assignments
predicted_scores_grid = lin_reg.predict(X_grid).reshape(hours_grid.shape)
```

```python
# Create a 3D plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Plot the original data points
ax.scatter(X[:, 0], X[:, 1], y, color='blue', s=100, label='Data Points')

# Plot the regression plane
ax.plot_surface(hours_grid, assignments_grid, predicted_scores_grid, color='orange', alpha=0.5)

# Labels and title
ax.set_xlabel('Hours Studied')
ax.set_ylabel('Assignments Completed')
ax.set_zlabel('Test Score')
ax.set_title('Linear Regression: Test Score Prediction')

plt.show()
```
